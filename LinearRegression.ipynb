```python
#import dependencies
from sklearn.linear_model import LinearRegression
import random
```

```python
#create 2 lists
feature_set = []
target_set = []
```

```python
#get the number of rows for the dataset
number_of_rows = 200

#limit possible value sin the dataset
random_number_limit = 2000
```

```python
#create the dataset

#create the feature dataset

for i in range(0, number_of_rows):
  x = random.randint(0,random_number_limit)
  y = random.randint(0,random_number_limit)
  z = random.randint(0,random_number_limit)

  #create linear function for target dataset
  function = (10*x)+(3*y)+(2*z)

  #append dataset to the lists
  feature_set.append([x,y,z])
  target_set.append(function)
```

```python
#create the linear regression model
model = LinearRegression()
model.fit(feature_set, target_set)
```

```python
#create the test dataset
test_set = [[8,10,0]]   #expected output = function(8,10,0) = 100
prediction = model.predict(test_set)
```

```python
print("Prediction:" + str(prediction) + "Coefficients:" + str(model.coef_))
```
